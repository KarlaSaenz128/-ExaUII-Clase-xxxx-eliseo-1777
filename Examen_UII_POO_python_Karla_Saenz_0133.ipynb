### Problema 48 : la clase "Batería"

#### sea interactivo --> los input
"Batería", con 3 atributos (marca, voltaje, capacidad), 2 funciones leerDatos(), mostrarDatos() y 2 objetos (pila1, pila2) no usar while true,if,for,listas y diccionario


![texto del vínculo](https://raw.githubusercontent.com/KarlaSaenz128/-ExaUII-Clase-xxxx-eliseo-1777/refs/heads/main/pilas.png)

```python
class Bateria:
    def __init__(self, marca, voltaje, capacidad):
        # Inicializamos los atributos
        self.marca = ""
        self.voltaje = ""
        self.capacidad = ""

    def leerDatos(self):
        # Captura de los 3 datos por teclado
        print("\n--- Ingresando Información de la Batería ---")
        self.marca = input("Ingrese la marca: ")
        self.voltaje = input("Ingrese el voltaje (ej. 12V): ")
        self.capacidad = input("Ingrese la capacidad (ej. 5000mAh): ")

    def mostrarDatos(self):
        # Exhibición de los 3 atributos
        print("Marca:", self.marca)
        print("Voltaje:", self.voltaje)
        print("Capacidad:", self.capacidad)

# --- Creación de Objetos (pila1 y pila2) ---

# Operaciones para la Pila 1
print("--- DATOS DE LA PILA 1 ---")
pila1 = Bateria("", "", "")
pila1.leerDatos()
print("-*-*-*-*")
pila1.mostrarDatos()
print("-*-*-*-*")

# Operaciones para la Pila 2
print("\n--- DATOS DE LA PILA 2 ---")
pila2 = Bateria("", "", "")
pila2.leerDatos()
print("-*-*-*-*")
pila2.mostrarDatos()
print("-*-*-*-*")

print("\nELABORADO por karla saenz0133")
```
